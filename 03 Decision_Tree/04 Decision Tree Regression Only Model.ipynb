```python
# Library
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
```

## Data Pre-Processing

```python
# Import Data
filename = 'https://gist.githubusercontent.com/nicolomarana/2345ec45ed67533ff6937857a444a450/raw/34d8524422bbc6103fec5064810ff9ccbe287bc2/housing'
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataframe = read_csv(filename, names=names)
dataframe.head()
```

```python
array = dataframe.values
print("----------------------------------")
print("Numero di array:", array.shape)
print("----------------------------------")
print("Array[0]:", array[0])
print("----------------------------------")
print("Primo elemento array[0]","CRIM", array[0][0])
print("----------------------------------")
print("Ultimo elemento array[0]","MEDV", array[0][-1])
print("----------------------------------")
```

```python
X = array[:,0:13]
print("----------------------------------")
print("Numero di Input", X.shape)
print("----------------------------------")
print("X[0]:", X[0])
print("----------------------------------")
print("Primo elemento X[0]","CRIM", X[0][0])
print("----------------------------------")
print("Ultimo elemento X[0]","LSTAT", X[0][-1])
print("----------------------------------", "\n")

Y = array[:,13]
print("----------------------------------")
print("Numero di Output", Y.shape)
print("----------------------------------")
print("Y[0]:", Y[0])
print("----------------------------------")
```

# Evaluation

```python
# K-fold Cross Validation
kfold = KFold(n_splits=10, random_state=7)

# Metrica di valutazione
scoring = 'neg_mean_squared_error'
```

```python
# Modello Linear Regression
model = DecisionTreeRegressor()
```

```python
results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print(results.mean())
```
