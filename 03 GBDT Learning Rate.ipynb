```
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import numpy
```

```
X, y = make_blobs(n_samples = 100, n_features = 2,
                       centers = 8, cluster_std = 1.3,
                       random_state = 4)
y = y % 2
plt.figure(dpi=150, edgecolor='w', facecolor='w')
plt.title('Sample binary classification problem with non-linearly separable classes')
plt.scatter(X[:,0], X[:,1], c=y,
           marker= 'o', s=50)
plt.show()
```

```
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```
# Random Choose LR

import random
lr = []

for i in range(10):
  lr.append(random.random())

#[0.001, 0.01, 0.1, 0.2, 0.5]

for lr in lr:
  clf = GradientBoostingClassifier(learning_rate=lr).fit(X_train, y_train)
  title = 'GBDT, complex binary dataset, default settings'
  print('Learning Rate:', lr, "Train Score:", round(clf.score(X_train, y_train),2),  "Test Score:", round(clf.score(X_test,y_test),2))
  
```

```
# Normal Choose

for lr in [0.001, 0.01, 0.05, 0.1, 0.2, 0.5]:
  clf = GradientBoostingClassifier(learning_rate=lr).fit(X_train, y_train)
  title = 'GBDT, complex binary dataset, default settings'
  print('Learning Rate:', lr, "Train Score:", round(clf.score(X_train, y_train),2),  "Test Score:", round(clf.score(X_test,y_test),2))
```
