```python
# Stochastic Gradient Boosting Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
```

## Data Pre-Processing

```python
# Import Data
filename = 'https://gist.githubusercontent.com/nicolomarana/728b2063e19ff4c6d56790a041693767/raw/f450082c7e2917545fda1a5bda10c4faad647ade/diabetes'
dataframe = read_csv(filename)
dataframe.head()
```

```python
array = dataframe.values
print("----------------------------------")
print("Numero di array:", array.shape)
print("----------------------------------")
print("Array[0]:","\n", array[0])
print("----------------------------------")
print("Primo elemento array[0]","Preg", array[0][0])
print("----------------------------------")
print("Ultimo elemento array[0]","Class", array[0][-1])
print("----------------------------------")
```

```python
X = array[:,0:8]
print("----------------------------------")
print("Numero di Input", X.shape)
print("----------------------------------")
print("X[0]:", X[0])
print("----------------------------------")
print("Primo elemento X[0]","Preg", X[0][0])
print("----------------------------------")
print("Ultimo elemento X[0]","Age", X[0][-1])
print("----------------------------------", "\n")

Y = array[:,8]
print("----------------------------------")
print("Numero di Output", Y.shape)
print("----------------------------------")
print("Y[0]:", Y[0])
print("----------------------------------")
```

# AdaBoost

```python
# Parameters & 10-Fold
seed = 7
num_trees = 100
kfold = KFold(n_splits=10, random_state=seed)
```

```python
# Model
model = GradientBoostingClassifier(n_estimators=num_trees, random_state=seed)
```

```python
# Evaluate
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```
